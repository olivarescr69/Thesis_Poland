```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Path to your CSV file
csv_path = r"C:\Users\MSI\OneDrive\1. BiG_dAtA_Master_Degree\thesis\Code\Data\Regression\Donations accepted by political parties.csv"

# Load the CSV
data = pd.read_csv(csv_path)
```

```python
# Select columns (definitive + optional)
selected_columns = [
    'Value', 'RegulatedEntityName', 'RegulatedEntityType', 'DonorStatus', 
    'DonationType', 'AcceptedDate', 'IsSponsorship', 'IsBequest', 
    'IsAggregation', 'AccountingUnitsAsCentralParty', 'IsIrishSource',
    'Postcode', 'IsReportedPrePoll', 'RegisterName'
]
data_selected = data[selected_columns]

# Step 1: Convert 'Value' to numeric for analysis
data_selected['Value'] = data_selected['Value'].replace('[\£,]', '', regex=True)
data_selected['Value'] = pd.to_numeric(data_selected['Value'], errors='coerce')

# Step 2: Analyze unique values for each column
print("Análisis de valores únicos por columna:")
for col in selected_columns:
    unique_count = data_selected[col].nunique()
    unique_values = data_selected[col].unique()
    print(f"\nColumna: {col}")
    print(f"Número de valores únicos: {unique_count}")
    print(f"Valores únicos (primeros 10 o todos si <10): {unique_values[:10]}")
    if unique_count > 10:
        print(f"(Mostrando solo primeros 10 de {unique_count} valores)")

# Step 3: Frequency analysis and visualization for categorical columns
cat_columns = [
    'RegulatedEntityName', 'RegulatedEntityType', 'DonorStatus', 
    'DonationType', 'IsIrishSource', 'IsReportedPrePoll', 'RegisterName'
]
for col in cat_columns:
    if data_selected[col].nunique() <= 20:  # Visualize only if manageable cardinality
        print(f"\nFrecuencias para {col}:")
        print(data_selected[col].value_counts())
        plt.figure(figsize=(10, 6))
        sns.countplot(data=data_selected, x=col, order=data_selected[col].value_counts().index)
        plt.title(f'Distribución de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"\nFrecuencias para {col} (top 5, alta cardinalidad):")
        print(data_selected[col].value_counts().head(5))

# Step 4: Frequency analysis for boolean columns
bool_columns = ['IsSponsorship', 'IsBequest', 'IsAggregation', 'AccountingUnitsAsCentralParty']
for col in bool_columns:
    print(f"\nFrecuencias para {col}:")
    print(data_selected[col].value_counts())
    plt.figure(figsize=(6, 4))
    sns.countplot(data=data_selected, x=col)
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

# Step 5: Analyze 'AcceptedDate' for temporal range
data_selected['AcceptedDate'] = pd.to_datetime(data_selected['AcceptedDate'], errors='coerce')
print("\nRango de AcceptedDate:")
print(f"Mínimo: {data_selected['AcceptedDate'].min()}")
print(f"Máximo: {data_selected['AcceptedDate'].max()}")
print("Frecuencias por año:")
print(data_selected['AcceptedDate'].dt.year.value_counts().sort_index())

# Step 6: Analyze 'Value' range
print("\nEstadísticas de Value:")
print(data_selected['Value'].describe())
plt.figure(figsize=(10, 6))
sns.histplot(data_selected['Value'].dropna(), bins=50, kde=True)
plt.title('Distribución de Value')
plt.xlabel('Monto de Donación (GBP)')
plt.ylabel('Frecuencia')
plt.xscale('log')  # Log scale for skewness
plt.show()

# Step 7: Analyze 'Postcode' cardinality
print("\nAnálisis de Postcode:")
print(f"Número de valores únicos: {data_selected['Postcode'].nunique()}")
print("Frecuencias (top 5):")
print(data_selected['Postcode'].value_counts().head(5))
```
